## Exploratory Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#Reading the csv file from month to month
months = ['01','02','03','04','05','06','07','08','09','10','11','12']
df_month = []
for month in months:
    df_month.append(pd.read_csv('2018'+month+'-fordgobike-tripdata.csv'))
df = pd.concat(df_month)

#Checking the dataset head
df.head()
```

```python
#Checking the number of rows
len(df)
```

```python
#Checking the entries and data types
df.info()
```

```python
#Correcting data types of objects

#Datetime
df['start_time'] =  pd.to_datetime(df['start_time'])
df['end_time'] =  pd.to_datetime(df['end_time'])

#Categorical
df['user_type'] = pd.Categorical(df.user_type)
df['bike_share_for_all_trip'] = pd.Categorical(df.bike_share_for_all_trip)

#String
df['bike_id'] = df.bike_id.apply(str)

#Printing datatypes
df.dtypes
```

```python
#Adding new column of day in week from end_date
df['day_in_week'] = df['start_time'].dt.dayofweek
```

## Data Visualization for Exploratory

### Missing Value

```python
#Declare the missing data
na_counts = df.isna().sum()

#Plot the missing data using barplot
base_color = sb.color_palette()[0]
fig, ax = plt.subplots(figsize=[30,4])
plt.title('Missing Values Barplot')
sb.barplot(na_counts.index.values, na_counts, color = base_color, ax=ax);
```

### Categorical Data

```python
#Creating plot table
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

#Plotting the pie chart for bike share
ax.axis('equal')
ax.pie(df['bike_share_for_all_trip'].value_counts(), labels = df['bike_share_for_all_trip'].unique(),autopct='%1.2f%%')
plt.title('Pie Chart for bike_share_for_all_trip in Data')

#Plotting the pie chart for user type
ax1 = fig.add_axes([1,0,1,1])
ax1.axis('equal')
ax1.pie(df['user_type'].value_counts(), labels = df['user_type'].unique(),autopct='%1.2f%%')
plt.title('Pie Chart for user_type in Data')

#Show plotting
plt.show()
```

### Trip Duration Distribution

```python
#Adjust the length of bin
bin_edges = np.arange(0, max(df['duration_sec'])+100, 100)

#Plot histogram
plt.title('Histogram for Trip Duration')
plt.hist(data=df, x='duration_sec',bins = bin_edges);
```

### This looks very bad! There's an outlier which have a very large value. We need to adjust our view so that we can see what we care the most, which is what the distribution looks like 

```python
#Create the length of bins
bin_edges = np.arange(0, 3000+100, 100)

#Plotting of histogram
plt.title('Histogram for Trip Duration')
plt.hist(data=df, x='duration_sec',bins = bin_edges);
```

### Looked better!

## Confusion Matrix of Features

```python
#Making new dummy df, dropping unneccesary columns for correlation matrix
corr_df = df.drop(columns=['bike_share_for_all_trip','user_type','start_station_id','start_station_name','end_station_id','end_station_name','end_station_latitude','start_station_latitude','end_station_longitude','start_station_longitude','bike_id'])
corr_df['start_day'] = df['start_time'].dt.day
corr_df['start_month'] = df['start_time'].dt.month
corr_df['end_month'] = df['end_time'].dt.month
corr_df['end_day'] = df['end_time'].dt.day
corr_df = corr_df.drop(columns=['start_time','end_time'])

#Generating Correlation Matrix
f = plt.figure(figsize=(8, 8))
plt.matshow(corr_df.corr(), fignum=f.number, cmap = 'Greens_r')
plt.xticks(range(corr_df.shape[1]), corr_df.columns, fontsize=14, rotation=45)
plt.yticks(range(corr_df.shape[1]), corr_df.columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16);
```

# Explanatory

### When are most trips taken in terms of time of day, day of the week, or month of the year?

**Have you ever wondered, when is the most trips taken at parts of the day, or parts of the week, or even parts of the year? Yeah, we too! Before we proceed to the findings, buckle up!**

```python
#Make variable for month data and create plot table
trips = df['start_time'].dt.hour
fig, ax = plt.subplots(figsize=[10,4])
plt.title('Distribution of Trips from hour to hour')

#Add constraints for highlighting of the max value
clrs = ['grey' if (x < max(trips.value_counts())) else 'red' for x in trips.value_counts(sort=False) ]

#Plotting of barplot
sb.barplot([i for i in range(24)], trips.value_counts(sort=False), ax=ax, palette=clrs);
plt.ylabel('count');
```

**Turns out, the trips peaked at 8-9am and 5-6pm! I'm sure the reason why this is the case because at 8-9am, people are going to work with their bike, and at 5pm, people are going back home, hence they need some vehicles and they decided to rent our bikes!**

```python
#Make variable for month data and create plot table
trips = df['day_in_week']
fig, ax = plt.subplots(figsize=[10,4])
plt.title('Distribution of Trips from Days of the Week')

#Add constraints for highlighting of the max value
clrs = ['grey' if (x < max(trips.value_counts())) else 'red' for x in trips.value_counts(sort=False) ]

#Plotting of barplot
sb.barplot([i for i in range(7)], trips.value_counts(sort=False), ax=ax, palette=clrs);
plt.ylabel('count')
plt.xticks(np.arange(7),['Monday', 'Tuesday', 'Wednesday','Thursday','Friday','Saturday','Sunday']);
```

**We find that, the trips is at peak on Tuesday and keeps on decreasing until Weekends. This might be because the user is mainly a working people. So they used the bike to go to work.**

```python
#Make variable for month data and create plot table
trips = df['start_time'].dt.day
fig, ax = plt.subplots(figsize=[10,4])
plt.title('Distribution of Trips from days to days')

#Add constraints for highlighting of the max value
clrs = ['grey' if (x < max(trips.value_counts())) else 'red' for x in trips.value_counts(sort=False) ]

#Plotting of barplot
sb.barplot([i for i in range(31)], trips.value_counts(sort=False), ax=ax, palette=clrs);
plt.ylabel('count');
```

**Now for this graph, we find that the most trips taken on 11 every month! The sample looks evenly distributed. However, we find a downfall at the end of the month. This might be because the salary isn't given yet, and they decided to save their money every single month.**

```python
#Make variable for month data and create plot table
trips = df['start_time'].dt.month
fig, ax = plt.subplots(figsize=[10,4])
plt.title('Distribution of Trips from month to month')

#Add constraints for highlighting of the max value
clrs = ['grey' if (x < max(trips.value_counts())) else 'red' for x in trips.value_counts(sort=False) ]

#Plotting of barplot
sb.barplot([i for i in range(12)], trips.value_counts(sort=False), ax=ax, palette=clrs);
plt.ylabel('count');
```

**Lastly, we find the most trips taken at September. September is the last month of summer, and the air feels really nice. This is why people love to bike on that weather!**

### How long does the average trip take? Does the above depend on if a user is a subscriber or customer?

**Next, we ask ourselves if the average trip taken is affected whether the user is a subscriber or customer. Surely, we find the answer!**

```python
#Creating dummy df for customer and subcriber
customer_df = df.query('user_type == "Customer" and duration_sec<3000')
subscriber_df = df.query('user_type == "Subscriber" and duration_sec<3000')

#Creating the table for plotting
fig, ax = plt.subplots(figsize=(8,8))
plt.title('Distribution of Trip Duration using Violin Plot')

#Plotting violin plot with the previous datasets
ax.boxplot(df.query('duration_sec<=3000')['duration_sec'],positions =[0]);
ax.boxplot(customer_df['duration_sec'],positions =[1]);
ax.boxplot(subscriber_df['duration_sec'],positions =[2]);

#Setting the labels
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(['Global', 'Customer', 'Subsriber'])
plt.show();
```

**We find that the member type is affeting the trip duration from this boxplot, we can see that the green line corresponds to the 50% percentile of the data. Next, we will find this findings in detail**

```python
#Creating the table for plotting
fig = plt.figure()

#Plotting the customer member data
base_color = sb.color_palette()
ax = fig.add_axes([0,1,1,1])
bin_edges = np.arange(0, 4000+50, 50)
plt.title('Histogram for Customer Member Duration')
ax.hist(data=customer_df, x='duration_sec',bins = bin_edges, color=base_color[1]);

#Plotting the subsriber member data
ax1 = fig.add_axes([1,1,1,1])
bin_edges = np.arange(0, 4000+50, 50)
plt.title('Histogram for Subscriber Member Duration')
ax1.hist(data=subscriber_df, x='duration_sec',bins = bin_edges, color=base_color[2]);

#Show plots
plt.show()
```

**Yes, we can see directly that the customer is more evenly distributed than the subsriber in this histogram**

```python
cus_mean = df.query('user_type == "Customer"')['duration_sec'].mean()
sub_mean = df.query('user_type == "Subscriber"')['duration_sec'].mean()
print('Average Trip: ',df['duration_sec'].mean(),'\n''Customer Mean:',cus_mean,'\n''Subscriber Mean:',sub_mean)
```

**Here we can see that the customer mean duration is 3 times larger than the subscriber! Amazing! This might be because the customer is someone who really need to use the bike in his everyday's life**
